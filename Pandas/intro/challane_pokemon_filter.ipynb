```python
import pandas as pd

df = pd.read_csv("csv_files/pokemon_data.csv")
```

Find all non-legendary Pokémon from Generation 1 whose Speed is higher than 100,
and show only their Name, Type 1, Speed, and Total stats — sorted by Speed (descending).

```python
df.head(3)

filter_df = df.loc[(~ df["Legendary"]) & 
                   (df["Generation"] == 1) & 
                   (df["Speed"] > 100)
                   ].copy()
filter_df["Total stats"] = filter_df["Total"] = filter_df['HP'] + filter_df['Attack'] + filter_df['Defense'] + filter_df['Sp. Atk'] + filter_df['Sp. Def'] + filter_df['Speed']
filter_df.sort_values(by=["Speed"], ascending= False)[["Name", "Type 1", "Speed", "Total stats"]]
```

Find all Pokémon that meet these criteria:

Are from Generation 2 or later
Have Attack greater than or equal to 120
Are not Legendary
Have at least one of their types as either "Fire" or "Dragon"

```python
filter_df = df.copy()
# filter_df.head(3)
filter_df = filter_df.loc[(filter_df["Generation"] >= 2) & 
              (filter_df["Attack"] >= 120) &
              (~filter_df["Legendary"])&
              (filter_df["Type 1"].isin(["Fire","Dragon"]) | filter_df["Type 2"].isin(["Fire","Dragon"]))
              ]
# filter_df.reset_index(inplace= True)
filter_df.reset_index(inplace= True, drop= True)
filter_df

```

Find all Pokémon that:

Are from Generation 2 or 3,
Have a Defense stat greater than 100,
Are not Legendary,
And have either "Psychic" or "Ghost" as their Type 1 or Type 2,
Then, add a new column called "Offense" which is the sum of Attack + Sp. Atk,
Finally, sort the results by "Offense" in descending order and show only these columns: Name, Type 1, Type 2, Defense, Offense

```python
df.head(3)
filter_df = df.loc[(df["Generation"].isin([2,3])) &
                   ~df["Legendary"] &
                   (df["Defense"] > 100) &
                   (df["Type 1"].isin(["Psychic", "Ghost"]) | df["Type 2"].isin(["Psychic","Ghost"])) 
                   ].copy()
filter_df["Offense"] = filter_df["Attack"] + filter_df["Sp. Atk"]
filter_df.sort_values(by="Offense", ascending = False)[["Name", "Type 1", "Type 2", "Defense", "Offense"]]
```

Challenge:
Find all Pokémon that meet these conditions:

Are from Generation 4 or later (Generation ≥ 4)
Have Speed greater than the average Speed of all Pokémon in the dataset
Are NOT Legendary
Have either Type 1 or Type 2 as "Water" or "Flying"
Create a new column called "Speed to Total Ratio" which is the Pokémon’s Speed divided by their Total stats
Sort the filtered Pokémon by "Speed to Total Ratio" in descending order
Return the columns: "Name", "Type 1", "Type 2", "Speed", "Total", "Speed to Total Ratio"

```python
df.columns


average_speed = df["Speed"].mean()
filter_df = df.loc[(df["Generation"] >= 4) & 
                   (df["Speed"] > average_speed)&
                   ~(df["Legendary"]) & 
                   (df["Type 1"].isin(["Water","Flying"]) | df["Type 2"].isin(["Water","Flying"]))
                   ].copy()
filter_df["Total"] = filter_df['HP'] + filter_df['Attack'] + filter_df['Defense'] + filter_df['Sp. Atk'] + filter_df['Sp. Def'] + filter_df['Speed']
filter_df["Speed to Total Ratio"] = (filter_df["Speed"] / filter_df["Total"]).round(4)
filter_df.reset_index(inplace= True, drop= True)
# filter_df.sort_values(by="Name", inplace= True)
filter_df.sort_values(by="Name")[["Name","Type 1","Type 2","Speed", "Total", "Speed to Total Ratio"]]
```

Using the Pokémon dataset df,

Filter all Pokémon that are Generation 1 to 5 and not Legendary.
Group these Pokémon by "Type 1" (primary type).
For each group, calculate:
The average Total stats (sum of all base stats: HP, Attack, Defense, Sp. Atk, Sp. Def, Speed)
The count of Pokémon in that type
Create a new DataFrame with columns:
["Type 1", "Average Total Stats", "Count"]
Sort this DataFrame by "Average Total Stats" in descending order.
Finally, print the top 3 types with the highest average total stats.

```python
filter_df = df.loc[(df["Generation"] <= 5) &
                   1 <= df["Generation"] &
                   ~df["Legendary"]
                   ].copy()

# grouping type 1
grouped = filter_df.groupby("Type 1")

# geting the average for each type 1
average_total_stats = grouped[["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]].mean()

# getting the average of all
average_total_stats["Average Total Stats"] = average_total_stats.sum(axis=1)

grouped_count = grouped.size().to_frame("Count")

# result_df = average_total_stats[["Average Total Stats"]].merge(grouped_count, left_index=True, right_index=True)
result_df = average_total_stats[["Average Total Stats"]].merge(grouped_count, left_index= True, right_index= True)

# making data ascending
result_df.sort_values(by="Average Total Stats", ascending=False)

result_df.head(3)
```
